```python
class Solution:
    def isPalindrome(self, s, left, right):
        # base case
        if left >= right:
            return True

        # mismatch case
        if s[left] != s[right]:
            return False

        # recursive call
        return self.isPalindrome(s, left + 1, right - 1)
sol = Solution()
result = sol.isPalindrome("nitin", 0, len("nitin") - 1)
print(result)
```
