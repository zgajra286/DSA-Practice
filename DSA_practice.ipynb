```python
class Solution:
    def isPalindrome(self, s, left, right):
        # base case
        if left >= right:
            return True

        # mismatch case
        if s[left] != s[right]:
            return False

        # recursive call
        return self.isPalindrome(s, left + 1, right - 1)
sol = Solution()
result = sol.isPalindrome("nitin", 0, len("nitin") - 1)
print(result)
```

```python
# QUICK SORT - Hoare Partition Style

arr = [4, 1, 7, 6, 3, 2, 8]

def partition(arr, low, high):

    pivot = arr[low]

    i = low + 1
    j = high

    while True:

        # move i right until wrong large element found
        while i <= j and arr[i] <= pivot:
            i += 1

        # move j left until wrong small element found
        while i <= j and arr[j] > pivot:
            j -= 1

        # pointers crossed
        if i > j:
            break

        # swap misplaced elements
        arr[i], arr[j] = arr[j], arr[i]

    # place pivot in correct position
    arr[low], arr[j] = arr[j], arr[low]

    return j


pivot_index = partition(arr, 0, len(arr)-1)

print(arr)
print("pivot index:", pivot_index)
```
